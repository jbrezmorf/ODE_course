# Overview of Common Mathematical Idioms

This notebook provides an overview of the most common idioms used in math lectures, especially for reading mathematical expressions and equations.

## 1. Fractions
- **$ \frac{a}{b} $** is often read as:
  - "a over b" (informal)
  - "a divided by b" (formal)
  - "the ratio of a to b"

Example: $ \frac{3}{5} $ is read as "three over five" or "three divided by five."

## 2. Exponents and Powers
- **$ a^b $** is typically read as:
  - "a to the power of b" (formal)
  - "a to the b" (more casual)
  - "a squared" (if $ b = 2 $)
  - "a cubed" (if $ b = 3 $)

Example: $ 2^3 $ is read as "two to the power of three" or "two cubed."

## 3. Roots
- **$ \sqrt{a} $** is read as:
  - "the square root of a"
- **$ \sqrt[n]{a} $** is read as:
  - "the nth root of a"
  - "the cube root of a" (if $ n = 3 $)

Example: $ \sqrt{9} $ is read as "the square root of nine," and $ \sqrt[3]{27} $ is read as "the cube root of twenty-seven."

## 4. Summation
- **$ \sum_{i=a}^{b} f(i) $** is read as:
  - "the sum from i equals a to b of f of i"
  - "summation from i equals a to b of f of i"

Example: $ \sum_{i=1}^{n} i^2 $ is read as "the sum from i equals 1 to n of i squared."

## 5. Products
- **$ ab $** is read as:
  - "a times b"
  - "a multiplied by b"

Example: $ 3 \times 4 $ is read as "three times four."

## 6. Differences
- **$ a - b $** is read as:
  - "a minus b"
  - "the difference between a and b"
  - "a, b difference"

## 7. Equation Operations
- **Subtraction**: When modifying an equation, one might say:
  - "subtract x from the equation"
  - "subtract x from both sides"

Example: "Subtract 2 from the equation to isolate the variable."

- **Substitution**:
  - "substitute x into the equation"
  - "plug x in"

Example: "Substitute $x = 3$ into the equation."

## 8. Term References
- Terms within an equation or expression are often referred to casually as "guys" or "terms."
  - "This guy here is multiplied by 2."
  - "Now we work with the second term."

## 12. Derivatives
- **$ \frac{d}{dx} f(x) $** is read as:
  - "the derivative of f of x with respect to x"
- **$ f'(x) $** is read as:
  - "f prime of x"

Example: $ \frac{d}{dx} x^2 $ is read as "the derivative of x squared with respect to x."

## 13. Integrals
- **$ \int_a^b f(x) \, dx $** is read as:
  - "the integral from a to b of f with respect to x"

Example: $ \int_0^1 x^2 \, dx $ is read as "the integral from 0 to 1 of x squared with respect to x."

## 14. Combinatorics
- **$ \binom{n}{k} $** is read as:
  - "n choose k"
  - "the combination of n taken k at a time"

Example: $ \binom{5}{2} $ is read as "five choose two."

## 16. Absolute Value
- **$ |a| $** is read as:
  - "the absolute value of a"

Example: $ |-3| $ is read as "the absolute value of negative three."

## 17. Set Notation
- **$ \{ a, b, c \} $** is read as:
  - "the set containing a, b, and c"
- **$ \subset $** is read as:
  - "is a subset of"
- **$ \cup $** is read as:
  - "union"
- **$ \cap $** is read as:
  - "intersection"
- **$ \emptyset $** is read as:
  - "the empty set"

Example: $ \{1, 2\} \cup \{2, 3\} $ is read as "the union of the set containing 1 and 2 and the set containing 2 and 3."

## 18. Logical Statements
- **$ \forall x $** is read as:
  - "for all x"
- **$ \exists x $** is read as:
  - "there exists an x"
- **$ P \implies Q $** is read as:
  - "P implies Q"
- **$ P \iff Q $** is read as:
  - "P if and only if Q", sometimes 

Example: $ \forall x \in \mathbb{R}, x^2 \geq 0 $ is read as "for all x in the real numbers, x squared is greater than or equal to zero."

## 19. Mini-Dictionary of Mathematical Idioms

- **"a times b"**: Used to describe multiplication. Example: "To simplify the equation, we calculate a times b."
- **"Subtract x from the equation"**: Refers to subtracting a term from both sides. Example: "Subtract 3 from both sides to isolate the variable."
- **"Substitute" or "Plug in"**: Refers to replacing a variable with a value. Example: "Substitute $x = 2$ into the equation."
- **"This guy here"**: A casual way to refer to a specific part of the equation. Example: "Now this guy here is what we need to simplify."
- **"Term"**: Refers to individual elements in an expression. Example: "Let's collect like terms on both sides."
